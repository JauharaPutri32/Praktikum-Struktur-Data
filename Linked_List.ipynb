```python
#LATIHAN#

print('1. Doubly Linked List')
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.prev = None

#Pembuatan node
n1 = Node('     Nanas')
n2 = Node('Apel')
n3 = Node('Durian')

class DoublyLinkedList:
    def __init__(self):
        self.head = None

    # Fungsi untuk menampilkan list
    def list(self):
        node = self.head
        while node:
            print(node.data, end=" <-> ")
            node = node.next
        print("None")

    #Penyisipan tengah
    def middle(self, mid_node, new_data):
        if mid_node is None:
            print("Node sebelumnya tidak ada.")
            return

        new_node = Node(new_data)
        new_node.next = mid_node.next
        mid_node.next = new_node
        new_node.prev = mid_node

        if new_node.next is not None:
            new_node.next.prev = new_node

    #Hapus dengan kata kunci
    def delete_key(self, key):
        curr = self.head
        if curr is None:
            return
        if curr.data == key:
            self.head = curr.next
            if self.head:
                self.head.prev = None
            curr = None
            return

        while curr and curr.data != key:
            curr = curr.next
        if curr is None:
            print(f"Node dengan kunci {key} tidak ditemukan.")
            return
        if curr.next:
            curr.next.prev = curr.prev
        if curr.prev:
            curr.prev.next = curr.next

dll = DoublyLinkedList()

dll.head = n1
n1.next = n2
n2.prev = n1
n2.next = n3
n3.prev = n2

print("  List Awal:")
dll.list()

dll.middle(dll.head.next, 'Kelengkeng')
print("\n  a) Penyisipan Tengah ('Kelengkeng'):")
dll.list()

dll.delete_key('Apel')
print("\n  b) Hapus dengan Kata Kunci ('Apel'):")
dll.list()

print('\n======================================================================')
print('2. Circular Linked List')
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

#Pembuatan node
n1 = Node(f'     (25)')
n2 = Node(50)
n3 = Node(75)

class CircularLinkedList:
    def __init__(self):
        self.head = None

    # Fungsi untuk menampilkan list
    def list(self):
        nodes = []
        curr = self.head
        while True:
            nodes.append(str(curr.data))
            curr = curr.next
            if curr == self.head:
                break
        print(" -> ".join(nodes) + " -> (kembali ke :" + str(self.head.data) + ")")

    #Penyisipan akhir
    def last(self, data):
        new_node = Node(data)
        if not self.head:
            self.head = new_node
            new_node.next = self.head
            return

        temp = self.head
        while temp.next != self.head:
            temp = temp.next

        temp.next = new_node
        new_node.next = self.head

    #Hapus akhir
    def delete_last(self):
        if not self.head:
            print("List kosong")
            return

        if self.head.next == self.head:
            self.head = None
            return
        temp = self.head

        while temp.next.next != self.head:
            temp = temp.next
        temp.next = self.head


cll = CircularLinkedList()

cll.head = n1
n1.next = n2
n2.next = n3
n3.next = cll.head

print("  List Awal:")
cll.list()

cll.last(100)
print("\n  a) Penyisipan Akhir (100):")
cll.list()

cll.delete_last()
print("\n  b) Hapus Akhir:")
cll.list()
```

```python
#Penyisipan node pada linked list

class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

#create node
n1 = Node("Januari")
n2 = Node("Februari")
n3 = Node("Maret")

class LinkedList:
  def __init__(self):
    self.headval = None

#print the linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

#create atribut add begining to update the new nodes next val to existing node
  def AddBegining(self,newdata):
    NewNode = Node(newdata)
    NewNode.nextval = self.headval
    self.headval = NewNode

#function to add node in the middle
  def AddInbetween(self,mid_node,newdata):
    if mid_node is None:
      print("The mentioned node is absent")
      return
    NewNode = Node(newdata)
    NewNode.nextval = mid_node.nextval
    mid_node.nextval = NewNode

#function to add newnode in the end
  def AddInEnd(self,newdata):
    NewNode = Node(newdata)
    if self.headval is None:
      self.headval = NewNode
      return
    last = self.headval
    while (last.nextval):
      last = last.nextval
    last.nextval = NewNode

Li = LinkedList()
Li.headval = n1

#link first Node to second node
Li.headval.nextval = n2

#link second Node to third node
n2.nextval = n3

Li.AddBegining("Start")
#add node after n1
Li.AddInbetween(n1,"Middle")
#add node in the end
Li.AddInEnd("The Last")
Li.listprint()

```

```python
#Menghapus node pada linked list

class Node:
  def __init__(self, dataval=None):
    self.dataval = dataval
    self.nextval = None

class LinkedList:
  def __init__(self):
    self.headval = None

#print the linked list
  def listprint(self):
    printval = self.headval
    while printval is not None:
      print (printval.dataval)
      printval = printval.nextval

#function to add node in the begining
  def AddBegining(self,newdata):
    NewNode = Node(newdata)
#update the new nodes next val to existing node
    NewNode.nextval = self.headval
    self.headval = NewNode

#function to remove node by key
  def RemoveNode(self, Removekey):
    HeadVal = self.headval
    if (HeadVal is not None):
      if (HeadVal.dataval == Removekey):
        self.headval = HeadVal.next
        HeadVal = None
        return
    while (HeadVal is not None):
      if HeadVal.dataval == Removekey:
        break
      prev = HeadVal
      HeadVal = HeadVal.nextval
    if (HeadVal == None):
      return
    prev.nextval = HeadVal.nextval
    HeadVal = None

#function to remove first node
  def RemoveFirst(self):
    afterhead = self.headval
    self.headval = afterhead.nextval

#function to remove last node
  def RemoveEnd(self):
    last = self.headval
    while (last is not None):
      if last.nextval == None:
        break
      prev = last
      last = last.nextval
    if (last == None):
      return
    prev.nextval = last.nextval
    last = None

#create linked list named Li
Li = LinkedList()

#add node
Li.AddBegining("April")
Li.AddBegining("Maret")
Li.AddBegining("Februari")
Li.AddBegining("Januari")
print("Sebelum dilakukan remove")
Li.listprint()
print("========================================================================")

#remove node by key = "Maret"
Li.RemoveNode("Maret")
print("Setelah dilakukan remove dengan kunci 'Maret'")
Li.listprint()
print("========================================================================")

#remove first node
Li.RemoveFirst()
print("Setelah dilakukan remove node pertama")
Li.listprint()
print("========================================================================")

#remove last node
Li.RemoveEnd()
print("Setelah dilakukan remove node terakhir")
Li.listprint()
```
